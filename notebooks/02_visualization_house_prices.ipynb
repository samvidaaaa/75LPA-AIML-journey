# Simple Linear Regression: Predicting House Prices
This notebook demonstrates basic linear regression to predict house prices based on features like size (sqft) and number of bedrooms. We also visualise the relationships between features and the target variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Features: Size (sqft), Bedrooms
X = pd.DataFrame({
    'Size_sqft': [1000, 1500, 2000, 2500],
    'Bedrooms': [2, 3, 3, 4]
})
```

```python
# Target: Price
y = pd.Series([200000, 250000, 300000, 400000], name = 'Price')
```

```python
plt.scatter(X['Size_sqft'], y)
plt.xlabel('Size (sqft)')
plt.ylabel('price')
plt.title('Size vs Price')
plt.show()

plt.scatter(X['Bedrooms'], y)
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.title('Bedrooms vs Price')
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)
print("Predicted prices:", predictions)
```

```python
# Coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Compare predicted vs actual
comparison = pd.DataFrame({'Actual': y, 'Predicted':predictions})
print(comparison)
```

## Insights

- **Coefficients** show the effect of each feature on house price:
  - Size (sqft): +100 per unit
  - Bedrooms: +25,000 per bedroom
- **Intercept**: 37,500  baseline price if all features are 0
- **Predictions vs Actuals**: Predicted prices are close to actual prices, showing the model captures the trend well.
